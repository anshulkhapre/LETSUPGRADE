```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
fmri.head(10)
```

```python
fmri.info()
```

```python
fmri.describe()
```

```python
fmri.shape
```

```python
fmri['timepoint']
```

```python
sns.scatterplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.relplot(x='signal',y='timepoint',data=fmri,hue='event',style='region')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',sort=True)
```

```python
sns.catplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.catplot(x='timepoint',y='signal',data=fmri,kind='box',hue='event',dodge=False)
```

```python
sns.catplot(x='timepoint',y='signal',data=fmri.sort_values('subject'),kind='boxen',hue='event',dodge=False)
```
